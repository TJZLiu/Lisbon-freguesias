### 1. Gap Statistic

The gap statistic is a statistical method for estimating the optimal number of clusters in a clustering algorithm such as k-means. The idea is to compare the sum of within-cluster distances for the actual data to the sum of within-cluster distances for data generated from a reference distribution with no clustering structure.

Here are the steps to compute the gap statistic:

Compute the sum of within-cluster distances for the actual data for each number of clusters from 1 to k_max.

Generate B reference datasets with the same size and dimensionality as the actual data, where each feature is rescaled to have the same mean and standard deviation as the corresponding feature in the actual data.

Cluster each reference dataset with the same number of clusters as in step 1, and compute the sum of within-cluster distances.

Compute the gap statistic for each number of clusters from 1 to k_max as the difference between the log of the sum of within-cluster distances for the reference data and the log of the sum of within-cluster distances for the actual data, multiplied by a factor that accounts for the variability of the reference data.

Select the number of clusters with the largest gap statistic as the optimal number of clusters.

The intuition behind the gap statistic is that if the actual data has no clustering structure, then the sum of within-cluster distances for the reference data should be similar to the sum of within-cluster distances for the actual data. On the other hand, if the actual data has a clustering structure, then the sum of within-cluster distances for the reference data should be larger than for the actual data. The gap statistic quantifies this difference and provides a way to select the number of clusters that maximizes the difference.

Note that the gap statistic is just one method for selecting the number of clusters in k-means, and it has its limitations. For example, it assumes that the data is generated from a reference distribution with no clustering structure, which may not be true in some cases. Additionally, the choice of reference distribution can affect the results. Therefore, it is always a good practice to combine the gap statistic with other methods and visualizations to validate the results.

#### define the gap statistic function

```python
# Gap Statistic for K means
def optimalK(data, nrefs=3, maxClusters=15):
    """
    Calculates KMeans optimal K using Gap Statistic 
    Params:
        data: ndarry of shape (n_samples, n_features)
        nrefs: number of sample reference datasets to create
        maxClusters: Maximum number of clusters to test for
    Returns: (gaps, optimalK)
    """
    gaps = np.zeros((len(range(1, maxClusters)),))
    resultsdf = pd.DataFrame({'clusterCount':[], 'gap':[]})
    for gap_index, k in enumerate(range(1, maxClusters)):
# Holder for reference dispersion results
        refDisps = np.zeros(nrefs)
# For n references, generate random sample and perform kmeans getting resulting dispersion of each loop
        for i in range(nrefs):
            
            # Create new random reference set
            randomReference = np.random.random_sample(size=data.shape)
            
            # Fit to it
            km = KMeans(k)
            km.fit(randomReference)
            
            refDisp = km.inertia_
            refDisps[i] = refDisp
# Fit cluster to original data and create dispersion
        km = KMeans(k)
        km.fit(data)
        
        origDisp = km.inertia_
# Calculate gap statistic
        gap = np.log(np.mean(refDisps)) - np.log(origDisp)
# Assign this loop's gap statistic to gaps
        gaps[gap_index] = gap
        
        resultsdf = resultsdf.append({'clusterCount':k, 'gap':gap}, ignore_index=True)
    return (gaps.argmax() + 1, resultsdf)
```

#### apply it

### 2. Elbow Method

The Elbow Method is a technique used to determine the optimal number of clusters in a clustering algorithm such as k-means. The method involves plotting the sum of squared distances (or inertia) between each data point and its closest centroid for different values of k (the number of clusters), and selecting the value of k where the decrease in sum of squared distances begins to level off or form an "elbow."

The intuition behind the Elbow Method is that as the number of clusters increases, the sum of squared distances will generally decrease, since each data point is closer to its assigned centroid. However, at some point, adding more clusters will only capture noise and unnecessary complexity in the data, without improving the overall quality of the clustering.

To apply the Elbow Method, we first run the k-means algorithm with different values of k and calculate the sum of squared distances for each k. We then plot these values against k and look for the "elbow point" where the decrease in sum of squared distances begins to level off. This point represents a trade-off between the goodness of fit (measured by the sum of squared distances) and the complexity of the model (measured by the number of clusters).

```python
# Elbow Method for K means
model = KMeans()
# k is range of number of clusters.
visualizer = KElbowVisualizer(model, k=(2,30), timings= True)
visualizer.fit(pca_df)        # Fit data to visualizer
visualizer.show()        # Finalize and render figure
```

##### As we can see in the plot, the result of 8 clusters is more efficient.

KElbowVisualizer is a visualization tool provided by the Yellowbrick library in Python that helps in selecting the optimal number of clusters for KMeans clustering. It is based on the elbow method, which suggests that the optimal number of clusters should be selected at the "elbow" point in a plot of the sum of squared distances between each point and its closest centroid versus the number of clusters.

The KElbowVisualizer creates a KMeans model for a range of cluster numbers and fits it to the data. It then computes the sum of squared distances for each cluster and generates a plot of the within-cluster sum of squared distances versus the number of clusters. The optimal number of clusters is selected at the "elbow" point, which is the point of inflection in the plot where the rate of decrease in the sum of squared distances starts to slow down.

### 3.Silhouette Coefficient

The Silhouette Coefficient is a measure of how well an individual data point is clustered, based on its similarity to other points in its own cluster compared to points in other clusters. The coefficient ranges from -1 to 1, with values closer to 1 indicating that the point is well-matched to its own cluster and poorly-matched to neighboring clusters, and values closer to -1 indicating the opposite.

The Silhouette Coefficient is commonly used to evaluate the quality of clustering results, especially for cases where the number of clusters is not known a priori. A high average Silhouette Coefficient across all data points suggests that the clustering algorithm has produced well-defined and distinct clusters, while a low average Silhouette Coefficient suggests that the clusters may be overlapping or poorly-separated.

The Silhouette Coefficient is calculated for each data point as follows:

Compute the average distance between the data point and all other data points in the same cluster. This is called the "intra-cluster distance."

Compute the average distance between the data point and all data points in the nearest neighboring cluster. This is called the "nearest-cluster distance."

Compute the Silhouette Coefficient as (nearest_cluster_distance - intra_cluster_distance) / max(nearest_cluster_distance, intra_cluster_distance).

A high Silhouette Coefficient for a data point indicates that it is well-matched to its own cluster and poorly-matched to neighboring clusters, while a low Silhouette Coefficient indicates the opposite. The average Silhouette Coefficient across all data points can be used to evaluate the overall quality of clustering.

```python
# Silhouette Score for K means
# k is range of number of clusters.
visualizer = KElbowVisualizer(model, k=(2,30),metric='silhouette', timings= True)
visualizer.fit(pca_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

##### As we can see in the plot, the results are not very satisfactory. The score is highest at the initial 2 clusters. 

If the Silhouette score is highest at the initial 2 clusters, it suggests that the data points are best separated into two distinct clusters based on the similarity of their features or attributes. The Silhouette score measures the similarity of a data point to its own cluster compared to other clusters, and ranges from -1 to 1, with higher values indicating better clustering.

In this case, a Silhouette score that is highest at the initial 2 clusters suggests that the data points have clear and distinct clusters that can be separated into two groups with little overlap between them. This can be useful for further analysis or decision-making, such as identifying distinct customer segments or product categories.

However, it is important to note that the Silhouette score alone may not be sufficient to determine the optimal number of clusters for a dataset, as it is just one of several metrics that can be used. It is often recommended to use multiple metrics and compare them to make an informed decision on the optimal number of clusters for a particular dataset.

### 4.Calinski-Harabasz Index

The Calinski-Harabasz score, also known as the Variance Ratio Criterion (VRC), is a metric for evaluating the quality of clustering results, specifically for K-means clustering. The score is based on the ratio of the between-cluster dispersion and the within-cluster dispersion, which is a measure of how much the clusters are separated from each other compared to how much they are compact. The score is defined as follows:

CH = (B/(k-1)) / (W/(n-k))

where B is the between-cluster dispersion, W is the within-cluster dispersion, k is the number of clusters, and n is the total number of data points.

The numerator of the CH score, B/(k-1), measures the variance of the means of the clusters, which indicates how well the clusters are separated from each other. The denominator, W/(n-k), measures the variance within the clusters, which indicates how compact the clusters are. The higher the CH score, the better the clustering result is.

```python
# Calinski Harabasz Score for K means
# k is range of number of clusters.
visualizer = KElbowVisualizer(model, k=(2,30),metric='calinski_harabasz', timings= True)
visualizer.fit(pca_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

##### As we can see in the plot, the results are not very satisfactory. The score is highest at the initial 2 clusters. 

If the Calinski-Harabasz score is highest at the initial 2 clusters, it suggests that the data points are best separated into two distinct clusters based on the variance of their features or attributes. The Calinski-Harabasz score is a metric that measures the ratio of the between-cluster variance to the within-cluster variance, and higher values indicate better clustering.

### 5.Davies-Bouldin Index

The Davies-Bouldin score is another metric for evaluating the quality of clustering results, specifically for K-means clustering. It measures the average similarity between each cluster and its most similar cluster, while also taking into account the average size of the clusters. The score is defined as follows:

DB = (1/k) * sum(max((s(i) + s(j)) / d(c(i), c(j))) )

where k is the number of clusters, c(i) and c(j) are the centroids of clusters i and j, s(i) and s(j) are the average distances between the points in clusters i and j and their respective centroid, and d(c(i), c(j)) is the distance between the centroids of clusters i and j.

The Davies-Bouldin score is a measure of the ratio of the within-cluster scatter to the between-cluster separation. A lower score indicates better clustering, with scores closer to zero indicating more compact and well-separated clusters.

```python
# Davies Bouldin score for K means
def get_kmeans_score(data, center):
    '''
    returns the kmeans score regarding Davies Bouldin for points to centers
    INPUT:
        data - the dataset you want to fit kmeans to
        center - the number of centers you want (the k value)
    OUTPUT:
        score - the Davies Bouldin score for the kmeans model fit to the data
    '''
    #instantiate kmeans
    kmeans = KMeans(n_clusters=center)
# Then fit the model to your data using the fit method
    model = kmeans.fit_predict(pca_df)
    
    # Calculate Davies Bouldin score
    score = davies_bouldin_score(pca_df, model)
    
    return score
scores = []
centers = list(range(2,30))
for center in centers:
    scores.append(get_kmeans_score(pca_df, center))
    
plt.plot(centers, scores, linestyle='--', marker='o', color='b');
plt.xlabel('K');
plt.ylabel('Davies Bouldin score');
plt.title('Davies Bouldin score vs. K');
```

##### As we can see in the plot, the results are not very satisfactory. The score is lowest at the initial 2 clusters. 

### 6.Bayesian information criterion

Bayesian Information Criterion (BIC) is a criterion for model selection among a finite set of models. It is based on the likelihood function and the number of parameters in the model, and is often used in statistical modeling to avoid overfitting.

In the context of clustering analysis, BIC can be used to determine the optimal number of clusters for a given dataset. The general idea is to fit multiple clustering models with different numbers of clusters and calculate their BIC scores. The model with the lowest BIC score is considered the best model, and its corresponding number of clusters is chosen as the optimal number of clusters.

The BIC score for a clustering model with k clusters can be calculated using the following formula:

BIC = -2 * log(L) + k * log(n)

where L is the likelihood function of the model, k is the number of parameters (i.e., the number of clusters), and n is the number of data points.

The BIC score penalizes models with more parameters, so it tends to favor simpler models that can explain the data well with fewer clusters. However, it also takes into account the likelihood of the data, so it can balance the trade-off between model complexity and model fit.

```python
# BIC for GMM
from sklearn.mixture import GaussianMixture
n_components = range(1, 30)
covariance_type = ['spherical', 'tied', 'diag', 'full']
score=[]
for cov in covariance_type:
    for n_comp in n_components:
        gmm=GaussianMixture(n_components=n_comp,covariance_type=cov)
        gmm.fit(pca_df)
        score.append((cov,n_comp,gmm.bic(pca_df)))
# Visualize the results
score = pd.DataFrame(score).rename(columns={0: 'covariance type',1:'cluster numbers',2: 'BIC score'})
score = pd.pivot_table(score, values='BIC score', index=['cluster numbers'], columns=['covariance type'])
score.plot(figsize=(15, 10))
```

##### As we can see in the plot, a larger number of clusters always leads to a better result in all types of covariance

Bayesian Information Criterion (BIC) is a criterion for model selection that tries to balance the goodness of fit of a model with the complexity of the model. In Gaussian Mixture Models (GMM), the covariance_type parameter is one of the factors that affects the complexity of the model. The different covariance_type options result in models with different numbers of parameters, and hence, different levels of complexity.

In the context of BIC, a more complex model (i.e., a model with more parameters) is penalized by adding a term that depends on the number of parameters in the model. Therefore, choosing the appropriate covariance_type is crucial for finding the best-fitting GMM model that balances the model's goodness of fit and its complexity.

For example, if the data has a spherical shape, spherical covariance type is appropriate, and it results in a simpler model than other covariance types. On the other hand, if the data has a complex shape with different variances for different features, diag or full covariance type may be more appropriate. The choice of covariance_type depends on the data and the specific problem at hand, and it can have a significant impact on the GMM clustering results and the model selection based on BIC.

When comparing different models using the Bayesian Information Criterion (BIC), a lower BIC score indicates a better model. The BIC score is a measure of the goodness of fit of the model, and it takes into account both the likelihood of the data given the model and the complexity of the model (i.e., the number of parameters in the model).

The BIC score penalizes models with more parameters, so a simpler model with fewer parameters will have a lower BIC score compared to a more complex model with more parameters. Therefore, when comparing different models, the model with the lowest BIC score is preferred as it is a better trade-off between goodness of fit and model complexity.

### Calculate each spatial weight clustering isoperimetric quotient (IPQ)

One very simple measure of geographical coherence involves the “compactness” of a given shape. The most common of these measures is the isoperimetric quotient [HHV93]. This compares the area of the region to the area of a circle with the same perimeter as the region. To obtain the statistic, we can recognize that the circumference of the circle  is the same as the perimeter of the region i , so Pi = 2πr . Then, the area of the isoperimetric circle is  Ac = π r2 = π(Pi/2π)2. Simplifying, we get: IPQ = Ai/Ac = 4πAi/Pi2
For this measure, more compact shapes have an IPQ closer to 1, whereas very elongated or spindly shapes will have IPQs closer to zero. For the clustering solutions, we would expect the IPQ to be very small indeed, since the perimeter of a cluster/region gets smaller the more boundaries that members share.
Computing this, then, can be done directly from the area and perimeter of a region:

```python
results = []
for cluster_type in w_methods:
    # compute the region polygons using a dissolve
    regions = df_pca[[cluster_type, "geometry"]].dissolve(by=cluster_type)
    # compute the actual isoperimetric quotient for these regions
    ipqs = (
        regions.area * 4 * np.pi / (regions.boundary.length ** 2)
    )
    # cast to a dataframe
    result = ipqs.to_frame(cluster_type)
    results.append(result)
# stack the series together along columns
ipqs = pd.concat(results, axis=1)
ipqs.describe()
```

### The Moran scores of all independent variables were calculated to determine their spatial autocorrelation

```python
w = weights.contiguity.Queen.from_dataframe(df_21)

All_varialbles21 = df_21.select_dtypes(include=['float']).columns.to_list()
# Set seed for reproducibility
np.random.seed(123456)
# Calculate Moran's I for each variable
mi_results = [
    Moran(df_21[variable], w) for variable in All_varialbles21
]
# Structure results as a list of tuples
mi_results = [
    (variable, res.I, res.p_sim)
    for variable, res in zip(All_varialbles21, mi_results)
]
# Display on table
table21 = pd.DataFrame(
    mi_results, columns=["Variable", "Moran's I", "P-value"]
).set_index("Variable")
table21.sort_values(by=["Moran's I"],ascending=False)
```

### Retain the independent variables with Moran score greater than 0.3

```python
# set the autocorrelation_variables in a list
autocorrelation_variables21 = table21[table21["Moran's I"] > 0.30].index.to_list()
```

### Visualization of the distribution of the variables in Lisbon 2021. Presenting data at the freguesia level

```python
FR21 = df_21.dissolve(by=['FR21'])

# the spatial distribution of each variable alone. 
f, axs = plt.subplots(nrows=14, ncols=2, figsize=(15, 50))
# Make the axes accessible with single indexing
axs = axs.flatten()
# Start a loop over all the variables of interest
for i, col in enumerate(autocorrelation_variables21):
    # select the axis where the map will go
    ax = axs[i]
    # Plot the map
    FR21.plot(
        column=col,
        ax=ax,
        scheme="Quantiles",
        linewidth=0,
        cmap="RdPu",
    )
    # Remove axis clutter
    ax.set_axis_off()
    # Set the axis title to the name of variable being plotted
    ax.set_title(col)
# Display the figure
plt.show()
```

### Determining the number of clusters by using KMeans

```python
# use sse coefficients and Elbow method to select the number of clusters 
sse = []
scores = []
for i in range(2, 25):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df_21[autocorrelation_variables21])
    scores.append(silhouette_score(df_21[autocorrelation_variables21], kmeans.labels_, metric='euclidean'))
    sse.append(kmeans.inertia_)
    
x = [i for i in range(2,25)]
 
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title('KMeans clustering SSE coefficients')
plt.xlabel('Number of clusters')
plt.plot(x, sse, 'o-')
 
plt.subplot(1,2,2)
plt.title('Average silhouette plot over clusters')
plt.xlabel('Number of clusters')
for i in range(2, 20):
    plt.text(i, scores[i-2], (str(round(scores[i-2], 2))))    
plt.plot(x, scores, 'o-')
```
